```python
from autogen import LLMConfig
import nest_asyncio
from autogen.agents.experimental import DeepResearchAgent
nest_asyncio.apply()


deep_research_llm_config = {
    "config_list": [{"api_type": "openai", "model": "gpt-4o-mini", "api_key": "API_KEY"}],
}

agent = DeepResearchAgent(
    name="DeepResearchAgent",
    llm_config=deep_research_llm_config,
    system_message="You can read papers on the internet and other sources to provide detailed information on the topic.",
)


message = "What was the impact of DeepSeek on stock prices and why?"
result = agent.run(
    message=message,
    tools=agent.tools,
    max_turns=2,
    user_input=False,
    summary_method="reflection_with_llm",
)

print(result.summary)
```
